```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepath in a variable
dataset_csv = "./dataset_immigrantstat.csv"
immigrant_stat_df = pd.read_csv(dataset_csv)
pd.options.display.float_format = '{:,.0f}'.format
immigrant_stat_df
```

```python
print(immigrant_stat_df.columns)
```

```python
# Create a DataFrame with columns to total: Before 2001, 2001 to 2005, 2006 to 2010, 
# 2011 to 2015, 2016 to 2021, Total Immigrants

immigrant_total_df = immigrant_stat_df[['Regions', 'Total_Immigrants',
                                        'Before_2001', '2001_to_2005', '2006_to_2010', 
                                         '2011_to_2015', '2016_to_2021']]
immigrant_total_df.head()
```

```python
# Region name
Region_name = immigrant_total_df.set_index(["Regions"]).groupby(["Regions"])
Region_name
```

```python
# Calculate the total number of Regions in Ontario
Region_count = len(immigrant_stat_df["Total_Immigrants"].unique())
Region_count
```

```python
# Calculate the total number of immigrants in Ontario before 2001 till 2021
immigrant_count = immigrant_stat_df["Total_Immigrants"].sum()
immigrant_count
```

```python
immigrant_total_df.select_dtypes(include=[np.number]).mean()
```

```python
# Sort the regions by highest immigrants` in descending order and display the top 5 rows.
top_regions = immigrant_total_df.sort_values("Before_2001", ascending = False)
top_regions.head()
```

```python
df = pd.DataFrame([['Toronto', 1457075, 362720, 326680, 324700, 391685], ['Ottawa - Gatineau', 144720, 33350, 37185, 40295, 58295], ['Ottawa - Gatineau (Ontario part)', 129360, 27750, 29730, 32050, 48170],
                   ['Hamilton', 113205, 18735, 19545, 19820, 26545 ], ['Kitchener - Cambridge - Waterloo', 70490, 15300, 16195, 16810, 27785]],
                  columns=['Regions', 'Before_2001', '2001_to_2005', '2006_to_2010', '2011_to_2015', '2016_to_2021'])

# view data
print(df)
  
# plot grouped bar chart
df.plot(x='Regions',
        kind='bar',
        stacked=False,
        title='Top 5 Regions By Year Group')
```
